
 <br>
DROPPING SOME NOT USEFUL FEATURES :<br>


```python
df = df.drop(['id', 'date', 'zipcode'], axis=1)
```

```python
print(df.columns)
```


 <br>
Defining Our Features And Labels : <br>


```python
X = df.drop('price', axis=1).values
y = df['price'].values
```


 <br>
TRAIN TEST SPLIT & SCALING OUR DATA :<br>


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("Shape of Features of X_train For Neural Network : ", X_train.shape)
```


<br>
CREATING A NEURAL NETWORK MODEL : <br>


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
```

```python
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse')
```


 <br>
TRAINING OUR MODEL :<br>


```python
model.fit(x=X_train, y=y_train,
          validation_data=(X_test, y_test), batch_size=128,
          epochs=400)
```

Checking The Plot for our losses :

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
plt.show()
```

```python
predictions = model.predict(X_test)
```

```python
print("y_test shape:", y_test.shape)
```

```python
print("predictions shape :", predictions.shape)
```

```python
predictions = predictions.reshape(6480, )
```

```python
df_comb = pd.DataFrame({'Actual Values': y_test, "Predicted Values": predictions})
print(df_comb.head(20))
```

```python
plt.scatter(y_test,predictions)
plt.show()
""" 
Evaluation :
"""
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,explained_variance_score
```

```python
print("Mean Absolute Error :", mean_absolute_error(y_test, predictions))
print('\n')
print("Mean Squared Error :", mean_squared_error(y_test, predictions))
print('\n')
print("Variance Score :", explained_variance_score(y_test, predictions))
```

```python
Errors=y_test-predictions
sns.distplot(Errors)
plt.show()
```
